# Introduction to Matplotlib (basic plots: line, scatter, bar)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Basic Plots:

## Creating a Simple Line Plot

In this task, we will create a simple line plot to visualize the relationship between two sets of data. Line plots are fundamental tools in data visualization, often used to track changes over periods of time or continuous variables. This task will demonstrate how to generate a basic line plot, which forms the foundation for more complex visualizations.

```python
years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
sales = [100, 120, 150, 180, 200, 220, 250]
plt.plot(years, sales)
plt.show()
```

## Plotting Multiple Lines on the Same Graph

In this task, we will plot multiple lines on the same graph. This approach is useful for comparing different data sets or tracking multiple variables simultaneously. By overlaying several lines, we can easily identify trends, correlations, and differences between the datasets. This task will showcase how to effectively combine multiple lines in a single plot.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.plot(years, sales)
plt.plot(years + 3, sales * 2)
plt.plot(years + 1, sales - 2)
plt.show()
```

## Adding Markers to a Line Plot

In this task, we will add markers to a line plot. Markers can highlight specific data points on the plot, making it easier to see individual values and trends. By using markers, we can enhance the readability and detail of the plot, making it more informative and engaging. This task will illustrate how to incorporate markers into a line plot.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.plot(years, sales, marker = "o")
plt.plot(years + 3, sales * 2, marker = "o")
plt.plot(years + 1, sales - 2, marker = "X")
plt.show()
```

## Customizing Line Styles and Colors

In this task, we will customize line styles and colors in a plot. Customizing the appearance of lines helps in differentiating between multiple data sets and making the plot visually appealing. We will explore various line styles, such as dashed or dotted lines, and apply different colors to each line. This task will demonstrate how to enhance the aesthetic and clarity of the plot through customization.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.plot(years, sales, "bX--")
plt.plot(years + 3, sales * 2, "rX-")
plt.plot(years + 1, sales - 2, "kX--")
plt.show()
```

# 2. Scatter Plots:

## Creating a Basic Scatter Plot

In this task, we will create a basic scatter plot to visualize the relationship between two numerical variables. Scatter plots are ideal for identifying patterns, correlations, and outliers within data. This task will demonstrate how to generate a simple scatter plot, providing a foundation for more advanced customization and analysis.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.scatter(years, sales)
plt.show()
```

## Customizing Marker Styles, Sizes, and Colors

In this task, we will customize the marker styles, sizes, and colors in a scatter plot. By adjusting these attributes, we can enhance the visual appeal and readability of the plot, making it easier to distinguish between different data points or categories. This task will show how to effectively use various marker properties to convey additional information and improve the plot's clarity.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.scatter(years, sales, color = "r", marker = "x")
plt.show()
```

## Plotting Multiple Scatter Plots on the Same Graph

In this task, we will plot multiple scatter plots on the same graph. Combining multiple scatter plots allows us to compare different data sets or groups within the same visual context. This method is useful for identifying differences and similarities between the datasets. This task will demonstrate how to overlay several scatter plots in a single graph.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.scatter(years, sales, color = "b", marker = 'x')
plt.scatter(years + 2, sales, color = "r", marker = 'o')
plt.scatter(years + 4, sales + -2, color = "g", marker = '*')
plt.show()
```

## Adding Labels and Titles to the Scatter Plot

In this task, we will add labels and titles to the scatter plot. Clear and informative labels and titles are essential for understanding the context and meaning of the plot. By providing descriptive axis labels and a concise title, we can ensure that the scatter plot effectively communicates the key insights from the data. This task will illustrate how to enhance the plot with appropriate labels and titles.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.scatter(years, sales, color = "r", marker = 'x')
plt.scatter(years + 2, sales, color = "b", marker = 'o')
plt.scatter(years + 4, sales + -2, color = "g", marker = '*')
plt.xlabel("Dependent Element")
plt.ylabel("Independen Element")
plt.title("Matplotlib.pyplot")
plt.show()
```

# 3. Bar Charts:

## Creating a Vertical Bar Chart

In this task, we will create a vertical bar chart to represent categorical data. Vertical bar charts are effective for comparing the quantities of different categories at a glance. Each bar's height corresponds to the value of the category it represents, making it easy to compare values. This task will demonstrate how to generate a basic vertical bar chart.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.bar(years, sales)
plt.show()
```

## Creating a Horizontal Bar Chart

In this task, we will create a horizontal bar chart to display categorical data. Horizontal bar charts are particularly useful when dealing with long category labels or when you want to emphasize the comparison between categories. Each bar's length corresponds to the value of the category it represents. This task will show how to create a horizontal bar chart.

## Customizing Bar Colors and Adding Edge Colors

In this task, we will customize the colors of the bars and add edge colors in a bar chart. Customizing bar colors can make the chart more visually appealing and help to distinguish between different categories or data series. Adding edge colors can enhance the clarity and definition of each bar. This task will illustrate how to enhance a bar chart by applying different colors and edge styles.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.barh(years, sales, color='skyblue', edgecolor='black')
plt.show()
```

# 4. Pie Charts:

## Creating a Basic Pie Chart

In this task, we will create a basic pie chart to represent the proportions of different categories within a dataset. Pie charts are a popular way to visualize parts of a whole, showing the relative sizes of data segments as slices of a pie. This task will demonstrate how to generate a simple pie chart, providing a clear overview of the data distribution.

```python
data = {
    'Brand': ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others'],
    'Market Share (%)': [35, 30, 15, 10, 10]
}
df = pd.DataFrame(data)
plt.pie(data["Market Share (%)"])
plt.show()
```

## Customizing Slice Colors and Adding a Legend

In this task, we will customize the colors of the slices in a pie chart and add a legend. Customizing slice colors enhances the visual appeal of the chart and makes it easier to distinguish between different categories. Adding a legend provides a key to understand what each slice represents. This task will show how to apply different colors to the slices and include a legend for better clarity and interpretation.

```python
data = {
    'Brand': ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others'],
    'Market Share (%)': [35, 30, 15, 10, 10]
}
df = pd.DataFrame(data)
mycolors = ["black", "hotpink", "#4CAF50", "b", "pink"]
plt.pie(data["Market Share (%)"], colors = mycolors, labels = data["Brand"])
plt.legend()
plt.show()
```

## Exploding a Slice of the Pie Chart

In this task, we will explode (pull out) a slice of the pie chart to highlight a specific category. Exploding a slice draws attention to a particular segment, making it stand out from the rest of the chart. This technique is useful for emphasizing important data points or categories. This task will illustrate how to explode a slice in a pie chart to highlight specific information.

```python
data = {
    'Brand': ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others'],
    'Market Share (%)': [35, 30, 15, 10, 10]
}
df = pd.DataFrame(data)
myexplode = [0.05, 0.04, 0, 0.09, 0]
mycolors = ["black", "hotpink", "#4CAF50", "b", "pink"]
plt.pie(data["Market Share (%)"], colors = mycolors, labels = data["Brand"], explode = myexplode)
plt.legend()
plt.show()
```

## Adding Percentage Labels to Slices

In this task, we will add percentage labels to the slices of a pie chart. Percentage labels provide precise information about the proportion each slice represents, making the chart more informative and easier to understand. By displaying percentages directly on the slices, viewers can quickly grasp the relative sizes of the categories. This task will demonstrate how to add percentage labels to enhance the informational value of a pie chart.

```python
data = {
    'Brand': ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others'],
    'Market Share (%)': [35, 30, 15, 10, 10]
}
df = pd.DataFrame(data)
myexplode = [0.05, 0.04, 0, 0.09, 0]
mycolors = ["black", "hotpink", "#4CAF50", "b", "pink"]
plt.pie(data["Market Share (%)"], colors = mycolors, labels = data["Brand"], explode = myexplode, autopct='%1.1f%%')
plt.legend()
plt.show()
```
