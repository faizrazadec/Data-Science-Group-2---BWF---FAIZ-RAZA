# 1. Subplots:

# Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).
## Visualizing Data with Multiple Plot Types

In this task, we aim to create a figure that showcases various types of plots, each providing unique insights into data representation. Specifically, we will generate a figure with a 2x2 grid of subplots, where each subplot contains a different type of plot. The plots we will include are:

- **Line Plot**: Ideal for showing trends over time or continuous data.
- **Scatter Plot**: Perfect for illustrating the relationship between two variables.
- **Bar Plot**: Useful for comparing categorical data.
- **Histogram**: Great for displaying the distribution of a dataset.

By combining these plots into a single figure, we can effectively compare and contrast different methods of data visuaization.


```python
years = [2018, 2019, 2020, 2021, 2022]
sales = [100, 120, 90, 110, 95]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(years, sales)
axes[0, 1].scatter(years, sales)
axes[1, 0].bar(years, sales)
axes[1, 1].hist(years, bins=10)

plt.tight_layout()
plt.show()
```

# Customize the subplots with titles, axis labels, and legends.
## Visualizing Data with Multiple Plot Types

In this task, we aim to create a figure that showcases various types of plots, each providing unique insights into data representation. Specifically, we will generate a figure with a 2x2 grid of subplots, where each subplot contains a different type of plot. The plots we will include are:

- **Line Plot**: Ideal for showing trends over time or continuous data.
- **Scatter Plot**: Perfect for illustrating the relationship between two variables.
- **Bar Plot**: Useful for comparing categorical data.
- **Histogram**: Great for displaying the distribution of a dataset.

By combining these plots into a single figure, we can effectively compare and contrast different methods of data visualization.


```python
years = [2018, 2019, 2020, 2021, 2022]
sales = [100, 120, 90, 110, 95]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(years, sales)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Sales')

axes[0, 1].scatter(years, sales)
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Sales')

axes[1, 0].bar(years, sales)
axes[1, 0].set_title('Bar Plot')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('Sales')

axes[1, 1].hist(years, bins=10)
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# 2. Histograms:

## Create a histogram for a given numerical data set. Customize the number of bins.

In this task, we will create a histogram to represent a given numerical dataset. A histogram is a graphical representation that organizes a group of data points into user-specified ranges. By customizing the number of bins, we can adjust the granularity of the data representation to uncover more detailed insights or broader trends. This task will help in understanding the distribution and frequency of the data values.

```python
x = np.random.normal(170, 10, 250)
y = np.random.normal(170, 10, 100)
plt.figure()
plt.hist(x, bins = 20)
plt.show()
```

## Plot multiple histograms on the same figure to compare different data sets.

In this task, we will plot multiple histograms on the same figure to compare different datasets. By overlaying these histograms, we can directly observe the similarities and differences between the distributions of multiple data sets. This approach is particularly useful when comparing the statistical properties of different data groups or tracking changes in the distribution over time or across categories.

```python
x = np.random.normal(170, 10, 250)
y = np.random.normal(170, 10, 100)
plt.figure()
plt.hist(x, bins = 20)
plt.hist(y, bins = 20)
plt.show()
```

## Normalizing Histograms and Adding Labels and Titles

In this task, we will normalize histograms to ensure that the areas under the histograms sum up to one. Normalizing the histograms allows for a fair comparison of datasets with different sizes. Additionally, we will add appropriate labels and titles to the histograms to make them more informative and easier to understand. Clear labeling and titling are essential for effective data communication, ensuring that viewers can quickly grasp the key messages from the visualizations.

```python
x = np.random.normal(170, 10, 250)
y = np.random.normal(170, 10, 100)
plt.figure()
plt.hist(x, bins = 20)
plt.hist(y, bins = 20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data 1 and Data 2')

plt.show()
```

# 3. Box Plots:

## Creating a Box Plot for a Given Numerical Dataset

In this task, we will create a box plot to represent a given numerical dataset. A box plot, also known as a whisker plot, provides a graphical summary of the distribution of a dataset. It displays the dataset's minimum, first quartile, median, third quartile, and maximum values, making it easy to identify the spread and any potential outliers. This visualization is useful for quickly understanding the central tendency and variability of the data.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.figure()
plt.boxplot(years)
plt.show()
```

## Generating Multiple Box Plots in a Single Figure

In this task, we will generate multiple box plots in a single figure to compare different datasets. By displaying several box plots side by side, we can effectively compare the distributions of multiple data sets. This method is particularly useful for identifying differences in medians, variability, and the presence of outliers across different groups or conditions.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.figure()
plt.boxplot(years)
plt.boxplot(years-10)
plt.show()
```

## Customizing Box Plots with Titles, Labels, and Colors

In this task, we will customize box plots by adding titles, labels, and changing the colors. Customizing these elements enhances the readability and interpretability of the plots. Titles provide context, labels describe the axes, and colors can be used to differentiate between multiple datasets or highlight specific aspects of the data. These customizations help to convey the intended message more clearly and make the visualizations more engaging.

```python
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024])
sales = np.array([100, 120, 150, 180, 200, 220, 250])
plt.figure()
plt.boxplot(years, boxprops=dict(color="blue"))
plt.boxplot(years-10, boxprops=dict(color="green"))
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data 1 and Data 2')
plt.show()
```
