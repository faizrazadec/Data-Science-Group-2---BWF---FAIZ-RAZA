# Introduction to Seaborn (visualization with Seaborn) 

### Importing Dependieces

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
df = pd.read_csv("university_enrollment_(Mini-project-01).csv")
df.head()
```

# 1. Scatter Plot
## Visualizing Relationships between Two Continuous Variables
A scatter plot is a useful tool for visualizing the relationship between two continuous variables. It plots each data point as a point on a coordinate grid, showing the relationship between the x and y variables. Scatter plots can help identify patterns, correlations, and outliers in the data.

```python
sns.scatterplot(data = df, x = "enrollment_count", y = "post_score")
plt.show()
```

# 2. Line PlotPlot
## Visualizing Trends and Patterns over Time or Space
A line plot is a graphical representation of data that shows trends and patterns over time or space. It connects a series of data points with lines, displaying the relationship between the x and y variables. Line plots are useful for showing changes, trends, and patterns inta across groups.


```python
sns.lineplot(data = df, x = "enrollment_count", y = "post_score")
plt.show()
```

# 3. Bar Plot
## Comparing Categorical Data across Groups
A bar plot is a graphical representation of data that compares categorical data across groups. It uses bars to represent the count or proportion of data points in each category, making it easy to compare and contrast data across groups. Bar plots are useful for showing differences and similarities between groups.

```python
sns.barplot(data = df, x = "year", y = "enrollment_count")
plt.show()
```

# 4. Count Plot
## Visualizing the Distribution of Categorical Data
A count plot is a graphical representation of data that shows the distribution of categorical data. It uses bars to represent the count of data points in each category, providing a quick overview of the data distribution. Count plots are useful for understanding the frequency and proportion of different categories in the data.

```python
sns.countplot(x="course_type", data=df)
plt.show()
```

# 5. Box Plot
## Visualizing the Distribution of Continuous Data
A box plot is a graphical representation of data that shows the distribution of continuous data. It displays the median, quartiles, and outliers of the data, providing a concise overview of the data distribution. Box plots are useful for comparing and contrasting the distribution of continuous data across groups.

```python
sns.boxplot(data = df, x = "year", y = "post_score")
plt.show()
```

# *Extra* Histogram (Hist Plot)
## Visualizing the Distribution of Continuous Data
A histogram, also known as a hist plot, is a graphical representation that organizes a group of data points into user-specified ranges. It shows the frequency distribution of a continuous variable by dividing the data into bins and plotting the number of data points that fall into each bin. Histograms are useful for understanding the underlying frequency distribution (shape) of a set of continuous data, as well as identifying patterns such as skewness, kurtosis, and the presence of outlier.


```python
sns.histplot(df, x = "course_type", bins = 10)
plt.show()
```

```python
sns.scatterplot(data = df, x = "enrollment_count", y = "post_score")
sns.set_style("whitegrid", {"font.family": "serif", "font.serif": ["Times New Roman"]})
plt.show()
```

# 2. Adding Titles and Labels
Adding titles and labels to your plots is essential for making them informative and easy to understand. Titles give an overview of what the plot represents, while axis labels describe the data presented along each axis. These elements can be added using matplotlib functions such as `plt.title()`, `plt.xlabel()`, and `plt.ylabel()`, enhancing the clarity and context of your visualizations.

```python
sns.scatterplot(data = df, x = "enrollment_count", y = "post_score")
sns.set_style("whitegrid", {"font.family": "serif", "font.serif": ["Times New Roman"]})
plt.title("This is Scatter Plot")
plt.xlabel("Enrollments")
plt.ylabel("Post Score")
plt.show()
```

# 3. Changing Palette
Changing the color palette in seaborn allows you to modify the color scheme used in your plots. This can help to differentiate data categories more effectively and make your plots visually appealing. The `sns.set_palette()` function provides various built-in palettes like "deep", "muted", "bright", "pastel", "dark", and "colorblind". You can also create custom color palettes to match your specific needs and preferences.

```python
sns.set_palette("viridis")
sns.scatterplot(data = df, x = "enrollment_count", y = "post_score")
sns.set_style("whitegrid", {"font.family": "serif", "font.serif": ["Times New Roman"]})
plt.title("This is Scatter Plot")
plt.xlabel("Enrollments")
plt.ylabel("Post Score")
plt.show()
```
